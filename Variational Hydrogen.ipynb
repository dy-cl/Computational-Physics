```python
from matplotlib import pyplot as plt
import numpy as np

def energy_func(c):
    E = (c**2)/2 - c

    return E

c = np.linspace(-1, 2, 100)
plt.plot(c, energy_func(c))
```

```python
from scipy.optimize import minimize

E_vals = []
c_vals = []

def minimise(c):
    E = (c**2)/2 - c

    E_vals.append(E)
    c_vals.append(c)

    print('c: ' + str(c))
    print('E: ' + str(E))

    return E

c = 5
result = minimize(minimise, c)

    
```

```python
plt.plot(c_vals, E_vals)
plt.xlabel('C')
plt.ylabel('E')
plt.title('Optimization Results')
plt.show()
```
