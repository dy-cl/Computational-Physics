```python
from scipy.optimize import minimize

c_values = []
E_values = []


def minimise(c):
    E = (c**2/2) - c

    c_values.append(c)
    E_values.append(E)

    print('C: ' + str(c))
    print('E: ' + str(E))
    print('\n')

    return E

c = 10

result = minimize(minimise, c)
```

```python
from matplotlib import pyplot as plt

plt.plot(c_values, E_values)
plt.xlabel('C')
plt.ylabel('E')
plt.title('Optimization Results')
plt.show()
```
