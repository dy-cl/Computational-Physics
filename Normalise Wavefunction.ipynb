An example wavefunction is given by,

$$ \psi = e^{-x^2/2} $$

and the probability density is given by

$$ \int_{-\infty}^{\infty} |\psi|^2 = \int_{-\infty}^{\infty} e^{-x^2/2}e^{-x^2/2}  = \int_{-\infty}^{\infty} e^{-x^2}$$

```python
import numpy as np

#Range of values
x = np.linspace(-10, 10, 1000) 

def psi(x):
    #Enter wavefunction form here
    return np.exp((-x**2)/2)

def psi_squared(x):
    #Probability density
    return np.abs(psi(x))**2 
```

However, in order for the probability density to be meaningful, the integral over all space must evaluate to 1, hence a normalisation constant is introduced such that

$$ \int_{-\infty}^{\infty} N\psi^*N\psi = N^2\int_{-\infty}^{\infty} |\psi|^2 = 1$$

First we check that the example $|\psi|^2$ is not already normalised.

```python
from scipy.integrate import quad

def integrand(x):
    return psi_squared(x)

#Integrate psi squared function over all space
integral = quad(integrand, -np.inf, np.inf)[0]
print("Integral result:", integral)
```

We see that for the Gaussian function the probability over all space is not 1, therefore it must be normalised. $N^2$ can be found by

$$ N^2 = 1 \div \int_{-\infty}^{\infty} |\psi|^2 $$

which when applied to the function should give 

$$ N^2\int_{-\infty}^{\infty} |\psi|^2 = 1 $$

```python
import matplotlib.pyplot as plt

N_squared = 1 / integral

def normalised_psi_squared(x):
    #Apply N_squared to wavefunction
    return N_squared * psi_squared(x)

normalised = normalised_psi_squared(x)

#Verify normalised function probability is equal to one
integral = quad(normalised_psi_squared, -np.inf, np.inf)[0]
print("Integral result:", integral)

#Plot all functions
plt.plot(x, psi(x), label = '$\psi$')
plt.plot(x, psi_squared(x), label = '$|\psi|^2$')
plt.plot(x, normalised, label = 'Normalised $|\psi|^2$')
plt.legend()
plt.show()
```
